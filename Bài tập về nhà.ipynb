```python

######################################################### BÀI TẬP VỀ NHÀ #############################################3

# BÀI 1

# BÀI 1: Trộn 2 mảng một chiều theo min
def tron_mang(a, b):
    """
    Trộn 2 mảng: lấy min của từng cặp phần tử tương ứng.
    Phần thừa của mảng dài hơn thì đưa vào cuối.
    """
    ket_qua = []
    min_len = min(len(a), len(b))

    # Lấy min của các phần tử tương ứng
    for i in range(min_len):
        ket_qua.append(min(a[i], b[i]))

    # Thêm phần thừa của mảng dài hơn
    if len(a) > len(b):
        ket_qua.extend(a[min_len:])
    else:
        ket_qua.extend(b[min_len:])

    return ket_qua


# Chạy Thử
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print("Mảng a:", a)
print("Mảng b:", b)
print("Kết quả sau khi trộn:", tron_mang(a, b))


```

```python
# BÀI 2

# CÂU A
def tinh_tong_tam_giac_tren_cheo_phu(A):
    """a. Tính tổng tam giác trên + đường chéo phụ (ma trận vuông)"""
    n = len(A)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:          # Điều kiện tam giác trên + chéo phụ
                tong += A[i][j]
    return tong
# CÂU B

def chuyen_am_thanh_tri_tuyet_doi(A):
    """b. Chuyển tất cả số âm thành trị tuyệt đối"""
    n = len(A)
    m = len(A[0])
    for i in range(n):
        for j in range(m):
            if A[i][j] < 0:
                A[i][j] = abs(A[i][j])
    return A

# CÂU C
def thay_chan_bang_x(A, x):
    """c. Thay tất cả số chẵn bằng x"""
    n = len(A)
    m = len(A[0])
    for i in range(n):
        for j in range(m):
            if A[i][j] % 2 == 0:
                A[i][j] = x
    return A

# CÂU D
def kiem_tra_toan_chan(A):
    """d. Kiểm tra ma trận có toàn số chẵn không"""
    for hang in A:
        for so in hang:
            if so % 2 != 0:
                return False
    return True
# CÂU E

def kiem_tra_doi_xung(A):
    """e. Kiểm tra ma trận vuông có đối xứng không"""
    n = len(A)
    for i in range(n):
        for j in range(n):
            if A[i][j] != A[j][i]:
                return False
    return True
# CÂU F

def kiem_tra_cheo_chinh_tang_dan(A):
    """f. Kiểm tra đường chéo chính có tăng dần không"""
    n = len(A)
    for i in range(1, n):
        if A[i][i] <= A[i-1][i-1]:
            return False
    return True

# CÂU G
def xuat_tam_giac_duoi_cheo_phu(A):
    """g. Xuất các phần tử tam giác dưới + chéo phụ"""
    n = len(A)
    print("Phần tử tam giác dưới đường chéo phụ:")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(A[i][j], end=" ")
        print()  # xuống dòng sau mỗi hàng

# CÂU H
def kiem_tra_cheo_phu_giam_dan(A):
    """h. Kiểm tra đường chéo phụ có giảm dần không"""
    n = len(A)
    for i in range(1, n):
        if A[i][n-1-i] >= A[i-1][n-1-(i-1)]:
            return False
    return True
```
