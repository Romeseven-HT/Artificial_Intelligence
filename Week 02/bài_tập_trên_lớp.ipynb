```python
import copy
import math
import numpy

X = "X"
O = "O"
EMPTY = None
user = None
ai = None


def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]



def player(board):
    count = 0
    for i in board:
        for j in i:
            if j is not None:
                count += 1
    return X if count % 2 == 0 else O



def actions(board):
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res


def result(board, action):
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board


# ===== WINNER =====
def get_horizontal_winner(board):
    for row in board:
        if row[0] is not None and row[0] == row[1] == row[2]:
            return row[0]
    return None


def get_vertical_winner(board):
    for i in range(3):
        if board[0][i] is not None and board[0][i] == board[1][i] == board[2][i]:
            return board[0][i]
    return None


def get_diagonal_winner(board):
    if board[0][0] is not None and board[0][0] == board[1][1] == board[2][2]:
        return board[0][0]

    if board[0][2] is not None and board[0][2] == board[1][1] == board[2][0]:
        return board[0][2]

    return None


def winner(board):
    return (get_horizontal_winner(board) or
            get_vertical_winner(board) or
            get_diagonal_winner(board))



def terminal(board):
    if winner(board) is not None:
        return True

    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True



def utility(board):
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0


# ===== MINIMAX =====
def maxValue(state):
    if terminal(state):
        return utility(state)

    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v


def minValue(state):
    if terminal(state):
        return utility(state)

    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v


def minimax(board):
    current_player = player(board)

    if current_player == X:
        best = -math.inf
        move = None
        for action in actions(board):
            val = minValue(result(board, action))
            if val > best:
                best = val
                move = action
        return move
    else:
        best = math.inf
        move = None
        for action in actions(board):
            val = maxValue(result(board, action))
            if val < best:
                best = val
                move = action
        return move


# ===== MAIN =====
if __name__ == "__main__":
    board = initial_state()
    ai_turn = False

    print("Choose a player (X/O):")
    user = input().upper()

    if user == "X":
        ai = "O"
    else:
        ai = "X"

    while True:
        game_over = terminal(board)
        playr = player(board)

        if game_over:
            win = winner(board)
            if win is None:
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {win} wins.")
            break

        else:
            if user != playr:
                print("AI move:")
                move = minimax(board)
                board = result(board, move)
                print(numpy.array(board))

            else:
                print("Enter the position to move (row,col)")
                i = int(input("Row: "))
                j = int(input("Col: "))

                if board[i][j] == EMPTY:
                    board = result(board, (i, j))
                    print(numpy.array(board))
                else:
                    print("Invalid move!")

Choose a player (X/O):
X
Enter the position to move (row,col)
Row: 0
Col: 1
[[None 'X' None]
 [None None None]
 [None None None]]
AI move:
[[None 'X' None]
 [None None None]
 [None 'O' None]]
Enter the position to move (row,col)
Row: 0
Col: 0
[['X' 'X' None]
 [None None None]
 [None 'O' None]]
AI move:
[['X' 'X' 'O']
 [None None None]
 [None 'O' None]]
Enter the position to move (row,col)
Row: 1
Col: 1
[['X' 'X' 'O']
 [None 'X' None]
 [None 'O' None]]
AI move:
[['X' 'X' 'O']
 [None 'X' None]
 [None 'O' 'O']]
Enter the position to move (row,col)
Row: 2
Col: 0
[['X' 'X' 'O']
 [None 'X' None]
 ['X' 'O' 'O']]
AI move:
[['X' 'X' 'O']
 [None 'X' 'O']
 ['X' 'O' 'O']]
Game Over: O wins.

import os, math

def GetWinner(board):
    """
    Returns the winner in the current board if there is one, otherwise it returns None.
    """
    # horizontal
    if board[0] == board[1] and board[1] == board[2]:
        return board[0]
    elif board[3] == board[4] and board[4] == board[5]:
        return board[3]
    elif board[6] == board[7] and board[7] == board[8]:
        return board[6]
    # vertical
    elif board[0] == board[3] and board[3] == board[6]:
        return board[0]
    elif board[1] == board[4] and board[4] == board[7]:
        return board[1]
    elif board[2] == board[5] and board[5] == board[8]:
        return board[2]
    # diagonal
    elif board[0] == board[4] and board[4] == board[8]:
        return board[0]
    elif board[2] == board[4] and board[4] == board[6]:
        return board[2]

def PrintBoard(board):
    """
    Clears the console and prints the current board.
    """
    os.system('cls' if os.name=='nt' else 'clear')
    print(f'''
    {board[0]}|{board[1]}|{board[2]}
    {board[3]}|{board[4]}|{board[5]}
    {board[6]}|{board[7]}|{board[8]}
    ''')

def GetAvailableCells(board):

    available = list()
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available

def minimax(position, depth, alpha, beta, isMaximizing):

    winner = GetWinner(position)
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth
    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"
            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return minEval

def FindBestMove(currentPosition, AI):

    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        moveVal = minimax(currentPosition, 0, -math.inf, +math.inf, False if AI == "X" else True)
        currentPosition[cell - 1] = cell
        if (AI == "X" and moveVal > bestVal):
            bestMove = cell
            bestVal = moveVal
        elif (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal
    return bestMove

def main():
    player = input("Play as X or O? ").strip().upper()
    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)]
    # X always starts first.
    currentTurn = "X"
    counter = 0
    while True:
        if currentTurn == AI:
            # NOTE: if the AI starts first, it'll always choose index 0 so to save time you could play it.
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        elif currentTurn == player:
            PrintBoard(currentGame)
            while True:
                humanInput = int(input("Enter Number: ").strip())
                if humanInput in currentGame:
                    currentGame[humanInput - 1] = player
                    currentTurn = AI
                    break
                else:
                    PrintBoard(currentGame)
                    print("Cell Not Available.")
        if GetWinner(currentGame) != None:
            PrintBoard(currentGame)
            print(f"{GetWinner(currentGame)} WON!!!")
            break
        counter += 1
        if GetWinner(currentGame) == None and counter == 9:
            PrintBoard(currentGame)
            print("Tie.")
            break

if __name__ == "__main__":
    main()


Play as X or O? X

    1|2|3
    4|5|6
    7|8|9

Enter Number: 2

    O|X|3
    4|5|6
    7|8|9

Enter Number: 5

    O|X|3
    4|X|6
    7|O|9

Enter Number: 7

    O|X|O
    4|X|6
    X|O|9

Enter Number: 6

    O|X|O
    O|X|X
    X|O|9

Enter Number: 9

    O|X|O
    O|X|X
    X|O|X

Tie.

# ====================== BÀI 1: MINIMAX CHO TIC-TAC-TOE 3x3 ======================
import math

# Khởi tạo bàn cờ
board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board(b):
    print("\nCột:  0   1   2")
    for idx, row in enumerate(b):
        print(f"Hàng {idx}: " + " | ".join(row))
        if idx < 2:
            print("    " + "-" * 11)
    print()

def is_moves_left(b):
    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                return True
    return False

def evaluate(b):
    # Kiểm tra hàng ngang và hàng dọc
    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] != ' ':
            return 10 if b[i][0] == 'O' else -10
        if b[0][i] == b[1][i] == b[2][i] != ' ':
            return 10 if b[0][i] == 'O' else -10

    # Kiểm tra đường chéo
    if b[0][0] == b[1][1] == b[2][2] != ' ':
        return 10 if b[0][0] == 'O' else -10
    if b[0][2] == b[1][1] == b[2][0] != ' ':
        return 10 if b[0][2] == 'O' else -10

    return 0

def minimax(b, depth, is_max):
    score = evaluate(b)

    if score == 10: return score - depth
    if score == -10: return score + depth
    if not is_moves_left(b): return 0

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    best = max(best, minimax(b, depth + 1, False))
                    b[i][j] = ' '
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    best = min(best, minimax(b, depth + 1, True))
                    b[i][j] = ' '
        return best

def find_best_move(b):
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                b[i][j] = 'O'
                move_val = minimax(b, 0, False)
                b[i][j] = ' '
                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)
    return best_move

# --- PHẦN ĐIỀU KHIỂN TRÒ CHƠI ---
def play_game():
    print("=== TIC-TAC-TOE 3x3 với Minimax ===")
    print("Bạn là 'X' (đi trước), AI là 'O'\n")

    while True:
        print_board(board)

        # 1. Lượt của người chơi
        while True:
            try:
                move = input("Nhập hàng và cột (vd: 0 1): ").split()
                r, c = int(move[0]), int(move[1])
                if board[r][c] == ' ':
                    board[r][c] = 'X'
                    break
                else:
                    print("Vị trí này đã có quân!")
            except:
                print("Nhập sai! Vui lòng nhập 2 số từ 0 đến 2.")

        if evaluate(board) == -10:
            print_board(board)
            print("Chúc mừng! Bạn đã thắng!")
            break
        if not is_moves_left(board):
            print_board(board)
            print("Hòa rồi!")
            break

        # 2. Lượt của AI
        print("AI đang suy nghĩ...")
        ai_move = find_best_move(board)
        if ai_move != (-1, -1):
            board[ai_move[0]][ai_move[1]] = 'O'

        if evaluate(board) == 10:
            print_board(board)
            print("AI thắng! thấy ghét")
            break
        if not is_moves_left(board):
            print_board(board)
            print("Hòa rồi!")
            break

if __name__ == "__main__":
    play_game()

=== TIC-TAC-TOE 3x3 với Minimax ===
Bạn là 'X' (đi trước), AI là 'O'


Cột:  0   1   2
Hàng 0:   |   |
    -----------
Hàng 1:   |   |
    -----------
Hàng 2:   |   |

Nhập hàng và cột (vd: 0 1): 0 0
AI đang suy nghĩ...

Cột:  0   1   2
Hàng 0: X |   |
    -----------
Hàng 1:   | O |
    -----------
Hàng 2:   |   |

Nhập hàng và cột (vd: 0 1): 2 2
AI đang suy nghĩ...

Cột:  0   1   2
Hàng 0: X | O |
    -----------
Hàng 1:   | O |
    -----------
Hàng 2:   |   | X

Nhập hàng và cột (vd: 0 1): 0 1
Vị trí này đã có quân!
Nhập hàng và cột (vd: 0 1): 2 1
AI đang suy nghĩ...

Cột:  0   1   2
Hàng 0: X | O |
    -----------
Hàng 1:   | O |
    -----------
Hàng 2: O | X | X

Nhập hàng và cột (vd: 0 1): 0 2
AI đang suy nghĩ...

Cột:  0   1   2
Hàng 0: X | O | X
    -----------
Hàng 1:   | O | O
    -----------
Hàng 2: O | X | X

Nhập hàng và cột (vd: 0 1): 1 0

Cột:  0   1   2
Hàng 0: X | O | X
    -----------
Hàng 1: X | O | O
    -----------
Hàng 2: O | X | X

Hòa rồi!

# ====================== BÀI 2: ALPHA-BETA PRUNING CHO TIC-TAC-TOE 3x3 ======================
import math

# Khởi tạo bàn cờ trống
board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board(b):
    print("\nCột:  0   1   2")
    for idx, row in enumerate(b):
        print(f"Hàng {idx}: " + " | ".join(row))
        if idx < 2: print("    " + "-" * 11)
    print()

def is_moves_left(b):
    return any(' ' in row for row in b)

def evaluate(b):
    # Kiểm tra hàng, cột và đường chéo
    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] != ' ':
            return 10 if b[i][0] == 'O' else -10
        if b[0][i] == b[1][i] == b[2][i] != ' ':
            return 10 if b[0][i] == 'O' else -10
    if b[0][0] == b[1][1] == b[2][2] != ' ':
        return 10 if b[0][0] == 'O' else -10
    if b[0][2] == b[1][1] == b[2][0] != ' ':
        return 10 if b[0][2] == 'O' else -10
    return 0

def minimax_alpha_beta(b, depth, is_max, alpha, beta):
    score = evaluate(b)


    if score == 10: return score - depth
    if score == -10: return score + depth
    if not is_moves_left(b): return 0

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    val = minimax_alpha_beta(b, depth + 1, False, alpha, beta)
                    b[i][j] = ' '
                    best = max(best, val)
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        return best
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    val = minimax_alpha_beta(b, depth + 1, True, alpha, beta)
                    b[i][j] = ' '
                    best = min(best, val)
                    beta = min(beta, best)
                    if beta <= alpha:
                        return best
        return best

def find_best_move_alpha_beta(b):
    best_val = -math.inf
    best_move = (-1, -1)


    alpha = -math.inf
    beta = math.inf

    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                b[i][j] = 'O'

                move_val = minimax_alpha_beta(b, 0, False, alpha, beta)
                b[i][j] = ' '

                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val
    return best_move

# --- VÒNG LẶP ĐIỀU KHIỂN CHÍNH ---
def main():
    print("=== TIC-TAC-TOE (ALPHA-BETA PRUNING) ===")
    print("AI (O) vs Bạn (X). Bạn đi trước!")

    while True:
        print_board(board)

        # Người chơi nhập tọa độ
        while True:
            try:
                inp = input("Lượt bạn (hàng cột, vd: 0 2): ").split()
                r, c = int(inp[0]), int(inp[1])
                if board[r][c] == ' ':
                    board[r][c] = 'X'
                    break
                else: print("Ô đã bị chiếm!")
            except: print("Nhập sai định dạng!")

        # Kiểm tra trạng thái sau khi người chơi đi
        if evaluate(board) == -10:
            print_board(board); print("Bạn thắng!"); break
        if not is_moves_left(board):
            print_board(board); print("Hòa!"); break

        # AI tìm nước đi
        print("AI đang tính toán cắt tỉa...")
        move = find_best_move_alpha_beta(board)
        if move != (-1, -1):
            board[move[0]][move[1]] = 'O'

        # Kiểm tra trạng thái sau khi AI đi
        if evaluate(board) == 10:
            print_board(board); print("AI thắng!"); break
        if not is_moves_left(board):
            print_board(board); print("Hòa!"); break

if __name__ == "__main__":
    main()


=== TIC-TAC-TOE (ALPHA-BETA PRUNING) ===
AI (O) vs Bạn (X). Bạn đi trước!

Cột:  0   1   2
Hàng 0:   |   |
    -----------
Hàng 1:   |   |
    -----------
Hàng 2:   |   |

Lượt bạn (hàng cột, vd: 0 2): 0 0
AI đang tính toán cắt tỉa...

Cột:  0   1   2
Hàng 0: X |   |
    -----------
Hàng 1:   | O |
    -----------
Hàng 2:   |   |

Lượt bạn (hàng cột, vd: 0 2): 2 2
AI đang tính toán cắt tỉa...

Cột:  0   1   2
Hàng 0: X | O |
    -----------
Hàng 1:   | O |
    -----------
Hàng 2:   |   | X

Lượt bạn (hàng cột, vd: 0 2): 2 1
AI đang tính toán cắt tỉa...

Cột:  0   1   2
Hàng 0: X | O |
    -----------
Hàng 1:   | O |
    -----------
Hàng 2: O | X | X

Lượt bạn (hàng cột, vd: 0 2): 0 2
AI đang tính toán cắt tỉa...

Cột:  0   1   2
Hàng 0: X | O | X
    -----------
Hàng 1:   | O | O
    -----------
Hàng 2: O | X | X

Lượt bạn (hàng cột, vd: 0 2): 1 0

Cột:  0   1   2
Hàng 0: X | O | X
    -----------
Hàng 1: X | O | O
    -----------
Hàng 2: O | X | X
```
