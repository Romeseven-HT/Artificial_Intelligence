```python
import math

SIZE = 5
MAX_DEPTH = 3

board = [[" " for _ in range(SIZE)] for _ in range(SIZE)]

# ===== HIỂN THỊ =====
def print_board():
    print("\nBàn cờ:")
    for i in range(SIZE):
        print(" | ".join(board[i]))
        if i < SIZE-1:
            print("-" * (SIZE*4-3))
    print()

# ===== KIỂM TRA =====
def check_winner():
    # hàng
    for i in range(SIZE):
        if board[i].count(board[i][0]) == SIZE and board[i][0] != " ":
            return board[i][0]

    # cột
    for j in range(SIZE):
        col = [board[i][j] for i in range(SIZE)]
        if col.count(col[0]) == SIZE and col[0] != " ":
            return col[0]

    # chéo chính
    if all(board[i][i] == board[0][0] and board[i][i] != " " for i in range(SIZE)):
        return board[0][0]

    # chéo phụ
    if all(board[i][SIZE-i-1] == board[0][SIZE-1] and board[i][SIZE-i-1] != " " for i in range(SIZE)):
        return board[0][SIZE-1]

    return None

def is_full():
    return all(board[i][j] != " " for i in range(SIZE) for j in range(SIZE))

# ===== MINIMAX =====
def minimax(isMax, depth):
    winner = check_winner()
    if winner == "X": return 1
    if winner == "O": return -1
    if is_full(): return 0

    if depth == MAX_DEPTH:
        return 0

    if isMax:
        best = -math.inf
        for i in range(SIZE):
            for j in range(SIZE):
                if board[i][j] == " ":
                    board[i][j] = "X"
                    best = max(best, minimax(False, depth+1))
                    board[i][j] = " "
        return best
    else:
        best = math.inf
        for i in range(SIZE):
            for j in range(SIZE):
                if board[i][j] == " ":
                    board[i][j] = "O"
                    best = min(best, minimax(True, depth+1))
                    board[i][j] = " "
        return best

# ===== AI CHỌN NƯỚC =====
def best_move(ai):
    human = "O" if ai == "X" else "X"

    #  1. Ưu tiên thắng ngay
    for i in range(SIZE):
        for j in range(SIZE):
            if board[i][j] == " ":
                board[i][j] = ai
                if check_winner() == ai:
                    board[i][j] = " "
                    return (i, j)
                board[i][j] = " "

    #  2. Chặn đối thủ thắng
    for i in range(SIZE):
        for j in range(SIZE):
            if board[i][j] == " ":
                board[i][j] = human
                if check_winner() == human:
                    board[i][j] = " "
                    return (i, j)
                board[i][j] = " "

    #  3. Minimax
    bestVal = -math.inf
    move = (-1, -1)

    for i in range(SIZE):
        for j in range(SIZE):
            if board[i][j] == " ":
                board[i][j] = ai
                val = minimax(ai == "O", 0)
                board[i][j] = " "
                if val > bestVal:
                    bestVal = val
                    move = (i, j)

    return move

# ===== GAME LOOP =====
def play():
    human = input("Chọn X hay O: ").upper()
    while human not in ["X", "O"]:
        human = input("Chọn lại (X/O): ").upper()

    ai = "O" if human == "X" else "X"
    turn = "X"

    print_board()

    while True:
        if turn == human:
            print("Lượt của bạn:")
            while True:
                try:
                    x = int(input("Nhập dòng (0-4): "))
                    y = int(input("Nhập cột (0-4): "))
                    if 0 <= x < SIZE and 0 <= y < SIZE and board[x][y] == " ":
                        board[x][y] = human
                        break
                    else:
                        print("Ô không hợp lệ!")
                except:
                    print("Nhập sai, thử lại!")

        else:
            print("Máy đang suy nghĩ...")
            x, y = best_move(ai)
            board[x][y] = ai

        print_board()

        winner = check_winner()
        if winner:
            print("Người thắng là:", winner)
            break

        if is_full():
            print("Hòa!")
            break

        turn = "O" if turn == "X" else "X"

# ===== CHẠY =====
play()
```

```python
# ====================== BÀI 4: ALPHA-BETA PRUNING CHO TIC-TAC-TOE NxN ======================
# Hỗ trợ mọi kích thước: 3x3, 5x5, 10x10, NxN
# AI chơi 'O' (Maximizer), Người chơi chơi 'X' (Minimizer)

class TicTacToeAlphaBeta:
    def __init__(self, size=5):
        self.size = size
        self.board = [[' ' for _ in range(size)] for _ in range(size)]
        self.player = 'X'   # Người chơi
        self.ai = 'O'       # Máy (AI)

    def print_board(self):
        """In bàn cờ dễ nhìn"""
        for i, row in enumerate(self.board):
            print(" | ".join(row))
            if i < self.size - 1:
                print("-" * (self.size * 4 - 1))
        print()

    def is_full(self):
        """Kiểm tra bàn cờ đã đầy chưa"""
        for row in self.board:
            if ' ' in row:
                return False
        return True

    def check_winner(self, player):
        """Kiểm tra người chơi có thắng không (hàng, cột, 2 đường chéo)"""
        # Kiểm tra hàng ngang
        for row in self.board:
            if all(cell == player for cell in row):
                return True

        # Kiểm tra cột dọc
        for col in range(self.size):
            if all(self.board[row][col] == player for row in range(self.size)):
                return True

        # Đường chéo chính
        if all(self.board[i][i] == player for i in range(self.size)):
            return True

        # Đường chéo phụ
        if all(self.board[i][self.size - 1 - i] == player for i in range(self.size)):
            return True

        return False

    def evaluate(self):
        """Đánh giá bàn cờ"""
        if self.check_winner(self.ai):
            return 10
        elif self.check_winner(self.player):
            return -10
        return 0

    # ====================== THUẬT TOÁN ALPHA-BETA PRUNING ======================
    def minimax_alpha_beta(self, depth, is_max, alpha, beta):
        score = self.evaluate()

        if score == 10:      # AI thắng
            return score - depth
        if score == -10:     # Người chơi thắng
            return score + depth
        if self.is_full():   # Hòa
            return 0

        if is_max:  # Lượt của AI (O)
            best = -1000
            for i in range(self.size):
                for j in range(self.size):
                    if self.board[i][j] == ' ':
                        self.board[i][j] = self.ai
                        best = max(best, self.minimax_alpha_beta(depth + 1, False, alpha, beta))
                        self.board[i][j] = ' '  # backtrack

                        alpha = max(alpha, best)
                        if beta <= alpha:      # CẮT TỈA
                            return best
            return best

        else:       # Lượt của người chơi (X)
            best = 1000
            for i in range(self.size):
                for j in range(self.size):
                    if self.board[i][j] == ' ':
                        self.board[i][j] = self.player
                        best = min(best, self.minimax_alpha_beta(depth + 1, True, alpha, beta))
                        self.board[i][j] = ' '  # backtrack

                        beta = min(beta, best)
                        if beta <= alpha:      # CẮT TỈA
                            return best
            return best

    def find_best_move(self):
        """Tìm nước đi tốt nhất cho AI bằng Alpha-Beta"""
        best_val = -1000
        best_move = (-1, -1)
        alpha = -1000
        beta = 1000

        print(f"AI đang suy nghĩ (Alpha-Beta) trên bàn cờ {self.size}x{self.size}...")

        for i in range(self.size):
            for j in range(self.size):
                if self.board[i][j] == ' ':
                    self.board[i][j] = self.ai
                    move_val = self.minimax_alpha_beta(0, False, alpha, beta)
                    self.board[i][j] = ' '

                    if move_val > best_val:
                        best_val = move_val
                        best_move = (i, j)

        return best_move


# ====================== CHẠY GAME ======================
if __name__ == "__main__":
    print("=== BÀI 4: ALPHA-BETA CHO TIC-TAC-TOE NxN ===")

    # Bạn có thể thay đổi kích thước ở đây
    size = int(input("Nhập kích thước bàn cờ (ví dụ: 3, 5, 7): "))

    game = TicTacToeAlphaBeta(size=size)

    print(f"\nBắt đầu trò chơi Tic-Tac-Toe {size}x{size}")
    print("Bạn chơi X (đánh trước), AI chơi O\n")

    while True:
        game.print_board()

        # Người chơi đánh
        while True:
            try:
                pos = int(input(f"Bạn (X) chọn vị trí (1 đến {size*size}): ")) - 1
                row = pos // size
                col = pos % size

                if 0 <= pos < size*size and game.board[row][col] == ' ':
                    game.board[row][col] = 'X'
                    break
                else:
                    print(" Vị trí không hợp lệ hoặc đã có quân! Chọn lại.")
            except:
                print(" Vui lòng nhập số hợp lệ!")

        # Kiểm tra kết thúc game
        if game.check_winner('X'):
            game.print_board()
            print(" BẠN (X) THẮNG!")
            break
        if game.is_full():
            game.print_board()
            print(" Hòa!")
            break

        # AI đánh
        best_move = game.find_best_move()
        if best_move != (-1, -1):
            row, col = best_move
            game.board[row][col] = 'O'

        # Kiểm tra kết thúc sau khi AI đánh
        if game.check_winner('O'):
            game.print_board()
            print(" AI (O) THẮNG!")
            break
        if game.is_full():
            game.print_board()
            print("  Hòa!")
            break

    print("\nCảm ơn bạn đã chơi!")
```
