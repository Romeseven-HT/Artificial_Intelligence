```python
import heapq

graph = {
    'A': {'B': 2, 'C': 4},
    'B': {'A': 2, 'D': 3, 'E': 1},
    'C': {'A': 4, 'E': 2, 'F': 5},
    'D': {'B': 3},
    'E': {'B': 1, 'C': 2, 'G': 3},
    'F': {'C': 5},
    'G': {}
}

h = {
    'A': 7, 'B': 5, 'C': 6, 'D': 9,
    'E': 3, 'F': 8, 'G': 0
}

def a_star(start, goal):
    open_list = []                    # priority queue
    heapq.heappush(open_list, (0, start))   # (f_score, node)

    g = {start: 0}                    # chi phí thực tế từ start
    parent = {start: None}            # để truy vết đường đi

    while open_list:
        current_f, current = heapq.heappop(open_list)

        if current == goal:
            break

        for neighbor, cost in graph[current].items():
            new_g = g[current] + cost

            if neighbor not in g or new_g < g[neighbor]:
                g[neighbor] = new_g
                f = new_g + h[neighbor]
                heapq.heappush(open_list, (f, neighbor))
                parent[neighbor] = current

    # Truy vết đường đi
    if goal not in parent:
        return None, None  # không tìm thấy đường đi

    path = []
    node = goal
    while node:
        path.append(node)
        node = parent[node]
    path.reverse()

    return path, g[goal]


# Chạy thử
path, cost = a_star('A', 'G')
print("Path:", " -> ".join(path))
print("Cost:", cost)
```

```python
import copy

def initial_state():
    return [[None, None, None],
            [None, None, None],
            [None, None, None]]

def player(board):
    x_count = sum(row.count('X') for row in board)
    o_count = sum(row.count('O') for row in board)
    return 'X' if x_count == o_count else 'O'

def actions(board):
    moves = set()
    for i in range(3):
        for j in range(3):
            if board[i][j] is None:
                moves.add((i, j))
    return moves

def result(board, action):
    i, j = action
    if board[i][j] is not None:
        raise Exception("INVALID MOVE")
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    lines = []

    # Hàng ngang
    lines.extend(board)

    # Cột dọc
    for j in range(3):
        lines.append([board[i][j] for i in range(3)])

    # Đường chéo chính
    lines.append([board[i][i] for i in range(3)])
    # Đường chéo phụ
    lines.append([board[i][2 - i] for i in range(3)])

    for line in lines:
        if line == ['X', 'X', 'X']:
            return 'X'
        if line == ['O', 'O', 'O']:
            return 'O'
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return all(cell is not None for row in board for cell in row)

def utility(board):
    w = winner(board)
    if w == 'X':
        return 1
    if w == 'O':
        return -1
    else:
        return 0

# ===================== MINIMAX =====================

def max_value(state):
    if terminal(state):
        return utility(state)
    v = float('-inf')
    for action in actions(state):
        v = max(v, min_value(result(state, action)))
    return v

def min_value(state):
    if terminal(state):
        return utility(state)
    v = float('inf')
    for action in actions(state):
        v = min(v, max_value(result(state, action)))
    return v

def minimax(board):
    if terminal(board):
        return None

    current_player = player(board)
    best_move = None

    if current_player == 'X':        # Người chơi (Maximizer)
        best_value = float('-inf')
        for action in actions(board):
            value = min_value(result(board, action))
            if value > best_value:
                best_value = value
                best_move = action
        return best_move
    else:                            # AI (O - Minimizer)
        best_value = float('inf')
        for action in actions(board):
            value = max_value(result(board, action))
            if value < best_value:
                best_value = value
                best_move = action
        return best_move

# ===================== GAME =====================

def print_board(board):
    for row in board:
        print([" " if cell is None else cell for cell in row])
    print()

def play_game():
    board = initial_state()

    while not terminal(board):
        print_board(board)

        if player(board) == 'X':          # Lượt người chơi
            try:
                move = input("Nhập nước đi (ví dụ: 1 2): ")
                i, j = map(int, move.split())
                i -= 1
                j -= 1

                if (i, j) not in actions(board):
                    print("Nước đi không hợp lệ!")
                    continue

                board = result(board, (i, j))
            except:
                print("Nhập sai! Vui lòng nhập theo dạng: hàng cột")
                continue

        else:                             # Lượt AI (O)
            print("AI đang suy nghĩ...")
            move = minimax(board)
            if move:
                board = result(board, move)

    # Kết thúc trò chơi
    print_board(board)
    w = winner(board)
    if w:
        print("Người thắng:", w)
    else:
        print("Hòa!")

# Chạy game
play_game()
```
