```python
# ====================== TRÒ CHƠI TIC-TAC-TOE 3x3 - NGƯỜI CHƠI VS AI ======================
# Bạn chơi 'X' (đánh trước), AI chơi 'O' (dùng Minimax)

board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board():
    for i, row in enumerate(board):
        print(" | ".join(row))
        if i < 2:
            print("-" * 9)
    print()

def is_moves_left():
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                return True
    return False

def evaluate():
    # Kiểm tra hàng, cột, đường chéo
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2]:
            if board[row][0] == 'O': return 10
            elif board[row][0] == 'X': return -10

    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col]:
            if board[0][col] == 'O': return 10
            elif board[0][col] == 'X': return -10

    if board[0][0] == board[1][1] == board[2][2]:
        if board[0][0] == 'O': return 10
        elif board[0][0] == 'X': return -10

    if board[0][2] == board[1][1] == board[2][0]:
        if board[0][2] == 'O': return 10
        elif board[0][2] == 'X': return -10

    return 0

def minimax(depth, is_max):
    score = evaluate()

    if score == 10:      # AI thắng
        return score - depth
    if score == -10:     # Người chơi thắng
        return score + depth
    if not is_moves_left():   # Hòa
        return 0

    if is_max:  # AI (O)
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    best = max(best, minimax(depth + 1, False))
                    board[i][j] = ' '
        return best
    else:       # Người chơi (X)
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    best = min(best, minimax(depth + 1, True))
                    board[i][j] = ' '
        return best

def find_best_move():
    best_val = -1000
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'
                move_val = minimax(0, False)
                board[i][j] = ' '

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)
    return best_move

def check_game_over():
    score = evaluate()
    if score == 10:
        print_board()
        print(" AI (O) THẮNG!")
        return True
    elif score == -10:
        print_board()
        print(" BẠN (X) THẮNG!")
        return True
    elif not is_moves_left():
        print_board()
        print("  Hòa!")
        return True
    return False

# ====================== CHẠY GAME ======================
print("=== TIC-TAC-TOE 3x3 ===")
print("Bạn chơi X (đánh trước) - AI chơi O")
print("Nhập vị trí theo số như sau:")
print(" 1 | 2 | 3 ")
print("---+---+---")
print(" 4 | 5 | 6 ")
print("---+---+---")
print(" 7 | 8 | 9 \n")

while True:
    print_board()

    # Người chơi đánh
    while True:
        try:
            pos = int(input("Bạn (X) chọn vị trí (1-9): ")) - 1
            row, col = divmod(pos, 3)
            if 0 <= pos <= 8 and board[row][col] == ' ':
                board[row][col] = 'X'
                break
            else:
                print(" Vị trí không hợp lệ hoặc đã có quân! Chọn lại.")
        except:
            print(" Vui lòng nhập số từ 1 đến 9!")

    if check_game_over():
        break

    # AI đánh
    print("AI đang suy nghĩ...")
    best_move = find_best_move()
    if best_move != (-1, -1):
        board[best_move[0]][best_move[1]] = 'O'

    if check_game_over():
        break

print("Cảm ơn bạn đã chơi!")
```

```python
# ====================== TRÒ CHƠI TIC-TAC-TOE 3x3 VỚI ALPHA-BETA PRUNING ======================
# Bạn chơi 'X' (đánh trước), AI chơi 'O' (sử dụng Alpha-Beta)

board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board():
    for i, row in enumerate(board):
        print(" | ".join(row))
        if i < 2:
            print("-" * 9)
    print()

def is_moves_left():
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                return True
    return False

def evaluate():
    # Kiểm tra hàng ngang
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2]:
            if board[row][0] == 'O': return 10
            elif board[row][0] == 'X': return -10

    # Kiểm tra cột dọc
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col]:
            if board[0][col] == 'O': return 10
            elif board[0][col] == 'X': return -10

    # Đường chéo chính
    if board[0][0] == board[1][1] == board[2][2]:
        if board[0][0] == 'O': return 10
        elif board[0][0] == 'X': return -10

    # Đường chéo phụ
    if board[0][2] == board[1][1] == board[2][0]:
        if board[0][2] == 'O': return 10
        elif board[0][2] == 'X': return -10

    return 0  # Hòa

# ====================== ALPHA-BETA PRUNING ======================
def minimax_alpha_beta(depth, is_max, alpha, beta):
    score = evaluate()

    if score == 10:      # AI thắng
        return score - depth
    if score == -10:     # Người chơi thắng
        return score + depth
    if not is_moves_left():   # Hòa
        return 0

    if is_max:  # Lượt AI (Maximizer - 'O')
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    best = max(best, minimax_alpha_beta(depth + 1, False, alpha, beta))
                    board[i][j] = ' '

                    alpha = max(alpha, best)
                    if beta <= alpha:   # CẮT TỈA
                        break
            if beta <= alpha:
                break
        return best

    else:       # Lượt người chơi (Minimizer - 'X')
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    best = min(best, minimax_alpha_beta(depth + 1, True, alpha, beta))
                    board[i][j] = ' '

                    beta = min(beta, best)
                    if beta <= alpha:   # CẮT TỈA
                        break
            if beta <= alpha:
                break
        return best

def find_best_move_alpha_beta():
    best_val = -1000
    best_move = (-1, -1)
    alpha = -1000
    beta = 1000

    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'
                move_val = minimax_alpha_beta(0, False, alpha, beta)
                board[i][j] = ' '

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move

def check_game_over():
    score = evaluate()
    if score == 10:
        print_board()
        print("🎉 AI (O) THẮNG!")
        return True
    elif score == -10:
        print_board()
        print("🎉 BẠN (X) THẮNG!")
        return True
    elif not is_moves_left():
        print_board()
        print("⚖️  Hòa!")
        return True
    return False

# ====================== CHẠY GAME ======================
print("=== TIC-TAC-TOE 3x3 với ALPHA-BETA PRUNING ===")
print("Bạn chơi X (đánh trước) - AI chơi O")
print("Nhập vị trí theo số như sau:")
print(" 1 | 2 | 3 ")
print("---+---+---")
print(" 4 | 5 | 6 ")
print("---+---+---")
print(" 7 | 8 | 9 \n")

while True:
    print_board()

    # Người chơi đánh
    while True:
        try:
            pos = int(input("Bạn (X) chọn vị trí (1-9): ")) - 1
            row, col = divmod(pos, 3)
            if 0 <= pos <= 8 and board[row][col] == ' ':
                board[row][col] = 'X'
                break
            else:
                print(" Vị trí không hợp lệ hoặc đã có quân! Chọn lại.")
        except:
            print(" Vui lòng nhập số từ 1 đến 9!")

    if check_game_over():
        break

    # AI đánh (sử dụng Alpha-Beta)
    print("AI đang suy nghĩ (Alpha-Beta)...")
    best_move = find_best_move_alpha_beta()
    if best_move != (-1, -1):
        board[best_move[0]][best_move[1]] = 'O'

    if check_game_over():
        break

print("Cảm ơn bạn đã chơi!")
```
