```python
# ====================== BÀI 1: MINIMAX CHO TIC-TAC-TOE 3x3 ======================
# AI là 'O', người chơi là 'X'.
board = [[' ' for _ in range(3)] for _ in range(3)]  # Bàn cờ 3x3

def print_board(b):
    for row in b:
        print(" | ".join(row))
        print("-" * 9)
    print()

def is_moves_left(b):
    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                return True
    return False

def evaluate(b):
    # Kiểm tra hàng, cột, đường chéo
    for row in range(3):
        if b[row][0] == b[row][1] == b[row][2]:
            if b[row][0] == 'O': return 10
            elif b[row][0] == 'X': return -10

    for col in range(3):
        if b[0][col] == b[1][col] == b[2][col]:
            if b[0][col] == 'O': return 10
            elif b[0][col] == 'X': return -10

    # Đường chéo chính
    if b[0][0] == b[1][1] == b[2][2]:
        if b[0][0] == 'O': return 10
        elif b[0][0] == 'X': return -10

    # Đường chéo phụ
    if b[0][2] == b[1][1] == b[2][0]:
        if b[0][2] == 'O': return 10
        elif b[0][2] == 'X': return -10

    return 0  # Hòa

def minimax(b, depth, is_max):
    score = evaluate(b)

    # Trường hợp cơ sở
    if score == 10:      # AI ('O') thắng
        return score - depth
    if score == -10:     # Người chơi ('X') thắng
        return score + depth
    if not is_moves_left(b):   # Hòa
        return 0

    if is_max:  # Lượt của AI (Maximizer - 'O')
        best = -1000
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    best = max(best, minimax(b, depth + 1, False))
                    b[i][j] = ' '  # backtrack
        return best
    else:       # Lượt của người chơi (Minimizer - 'X')
        best = 1000
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    best = min(best, minimax(b, depth + 1, True))
                    b[i][j] = ' '  # backtrack
        return best

def find_best_move(b):
    best_val = -1000
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                b[i][j] = 'O'
                move_val = minimax(b, 0, False)
                b[i][j] = ' '

                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val

    return best_move

print("=== TIC-TAC-TOE 3x3 với Minimax ===")
print("Bạn chơi X, AI chơi O\n")

```

```python
# ====================== BÀI 2: ALPHA-BETA PRUNING CHO TIC-TAC-TOE 3x3 ======================

def minimax_alpha_beta(b, depth, is_max, alpha, beta):
    score = evaluate(b)

    if score == 10:      # AI thắng
        return score - depth
    if score == -10:     # Người chơi thắng
        return score + depth
    if not is_moves_left(b):
        return 0

    if is_max:  # Maximizer (AI - 'O')
        best = -1000
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    best = max(best, minimax_alpha_beta(b, depth + 1, False, alpha, beta))
                    b[i][j] = ' '

                    alpha = max(alpha, best)
                    if beta <= alpha:   # CẮT TỈA
                        break
            if beta <= alpha:
                break
        return best

    else:       # Minimizer (Người chơi - 'X')
        best = 1000
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    best = min(best, minimax_alpha_beta(b, depth + 1, True, alpha, beta))
                    b[i][j] = ' '

                    beta = min(beta, best)
                    if beta <= alpha:   # CẮT TỈA
                        break
            if beta <= alpha:
                break
        return best

def find_best_move_alpha_beta(b):
    best_val = -1000
    best_move = (-1, -1)
    alpha = -1000
    beta = 1000

    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                b[i][j] = 'O'
                move_val = minimax_alpha_beta(b, 0, False, alpha, beta)
                b[i][j] = ' '

                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val

    return best_move

print("Alpha-Beta Pruning đã được cài đặt .")
```
