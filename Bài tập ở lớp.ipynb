```python
# Bài 13
def tron_mang_tong(a, b):
    """
    Trộn 2 mảng: phần tử mới = tổng của 2 phần tử tương ứng
    Phần thừa của mảng nào thì đưa hết vào cuối
    """
    ket_qua = []
    min_len = min(len(a), len(b))

    # Cộng từng cặp phần tử
    for i in range(min_len):
        ket_qua.append(a[i] + b[i])

    # Thêm phần thừa
    if len(a) > len(b):
        ket_qua.extend(a[min_len:])
    else:
        ket_qua.extend(b[min_len:])

    return ket_qua


# Chạy Thử
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print(tron_mang_tong(a, b))
```

```python
# bài 14
import random

def tao_ma_tran_ngau_nhien(m, n, min_val=1, max_val=100):
    """a. Tạo ma trận ngẫu nhiên"""
    return [[random.randint(min_val, max_val) for _ in range(n)] for _ in range(m)]


def xuat_dong_k(A, k):
    """b. Xuất dòng k (k bắt đầu từ 0)"""
    if 0 <= k < len(A):
        print(f"Dòng {k}:", A[k])
    else:
        print("Dòng không tồn tại!")


def xuat_cot_k(A, k):
    """c. Xuất cột k"""
    if 0 <= k < len(A[0]):
        print(f"Cột {k}:", [hang[k] for hang in A])
    else:
        print("Cột không tồn tại!")


def tim_dong_tong_lon_nhat(A):
    """d. Tìm dòng có tổng lớn nhất"""
    tong_cac_dong = [sum(hang) for hang in A]
    max_tong = max(tong_cac_dong)
    dong_max = tong_cac_dong.index(max_tong)
    print(f"Dòng {dong_max} có tổng lớn nhất = {max_tong}")
    return dong_max


def tim_cot_tich_nho_nhat(A):
    """e. Tìm cột có tích nhỏ nhất"""
    tich_cac_cot = []
    for j in range(len(A[0])):
        tich = 1
        for i in range(len(A)):
            tich *= A[i][j]
        tich_cac_cot.append(tich)

    min_tich = min(tich_cac_cot)
    cot_min = tich_cac_cot.index(min_tich)
    print(f"Cột {cot_min} có tích nhỏ nhất = {min_tich}")
    return cot_min


def xuat_dong_chan_cot_le(A):
    """f. Xuất phần tử thuộc dòng chẵn + cột lẻ"""
    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(len(A)):
        if i % 2 == 0:          # dòng chẵn
            for j in range(len(A[0])):
                if j % 2 == 1:  # cột lẻ
                    print(A[i][j], end=" ")
            print()


def trung_binh_dong_le(A):
    """g. Trung bình cộng các phần tử chẵn thuộc dòng lẻ"""
    tong = 0
    dem = 0
    for i in range(1, len(A), 2):   # dòng lẻ: 1,3,5,...
        for so in A[i]:
            if so % 2 == 0:         # số chẵn
                tong += so
                dem += 1
    return tong / dem if dem > 0 else 0


def trung_binh_bien(A):
    """h. Trung bình cộng các phần tử thuộc biên"""
    m, n = len(A), len(A[0])
    tong = 0
    dem = 0

    # Biên trên + biên dưới
    for j in range(n):
        tong += A[0][j]
        tong += A[m-1][j]
        dem += 2
    # Biên trái + biên phải (không tính góc 2 lần)
    for i in range(1, m-1):
        tong += A[i][0]
        tong += A[i][n-1]
        dem += 2

    return tong / dem if dem > 0 else 0


def trung_binh_khong_thuoc_bien(A):
    """i. Trung bình các phần tử không thuộc biên"""
    m, n = len(A), len(A[0])
    tong = 0
    dem = 0
    for i in range(1, m-1):
        for j in range(1, n-1):
            tong += A[i][j]
            dem += 1
    return tong / dem if dem > 0 else 0
```

```python
# bài 15
# a. Xây dựng cấu trúc SinhVien
class SinhVien:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = ma_sv          # Chuỗi 10 ký tự
        self.ten = ten              # Chuỗi tối đa 20 ký tự
        self.nam_sinh = nam_sinh    # Số nguyên
        self.diem_tb = diem_tb      # Số thực

    def __str__(self):
        return f"{self.ma_sv} - {self.ten} - {self.nam_sinh} - {self.diem_tb:.2f}"


# Tạo danh sách sinh viên mẫu
def tao_danh_sach_sinh_vien():
    ds = []
    ds.append(SinhVien("02DH0001", "Nguyen Van Lan", 2005, 8.5))
    ds.append(SinhVien("02DH0002", "Tran Thi Phan", 2004, 7.0))
    ds.append(SinhVien("02DH0003", "Le Van An", 2006, 9.2))
    ds.append(SinhVien("02DH0004", "Pham Lan Huong", 2005, 6.8))
    ds.append(SinhVien("02DH0005", "Hoang Thi Lan", 2003, 5.5))
    return ds


# Các hàm yêu cầu
def dem_sv_du_dieu_kien_len_lop(ds):
    """c. Đếm sinh viên có điểm TB >= 5"""
    dem = sum(1 for sv in ds if sv.diem_tb >= 5)
    print(f"Số sinh viên đủ điều kiện lên lớp: {dem}")
    return dem


def xuat_sv_20_tuoi(ds):
    """d. Xuất sinh viên 20 tuổi"""
    nam_hien_tai = 2026   # Giả sử năm hiện tại
    print("Sinh viên 20 tuổi:")
    for sv in ds:
        if nam_hien_tai - sv.nam_sinh == 20:
            print(sv)


def dem_sv_he_DH(ds):
    """e. Đếm sinh viên hệ Đại học (mã chứa 'DH' ở vị trí 2,3)"""
    dem = sum(1 for sv in ds if len(sv.ma_sv) >= 4 and sv.ma_sv[2:4] == "DH")
    print(f"Số sinh viên hệ Đại học: {dem}")
    return dem


def dem_sv_ten_Lan(ds):
    """f. Đếm sinh viên có tên chứa 'Lan'"""
    dem = sum(1 for sv in ds if "Lan" in sv.ten)
    print(f"Số sinh viên có tên Lan: {dem}")
    return dem


def dem_sv_ho_Phan(ds):
    """g. Đếm sinh viên có họ 'Phan'"""
    dem = sum(1 for sv in ds if sv.ten.split()[0] == "Phan" or sv.ten.split()[0] == "Phạm")
    print(f"Số sinh viên họ Phan: {dem}")
    return dem


```
